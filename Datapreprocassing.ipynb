Handling missing value in dataset


```python
#replace nan value with (mean,mode,median)
df['Age']=pd.to_numeric(df['Age'],errors='coerce')
df['Age'].fillna(df['Age'].mean(),inplace=True)
df.head()
```

```python
df
```

```python
#drop row with nan values / listwise deletion
df1= df.dropna()
df1
```

```python
#fill missing values using k nearest neighbour
from sklearn.impute import KNNImputer
knn_imputer=KNNImputer(n_neighbors=3)
df['Age']=pd.to_numeric(df['Age'],errors='coerce')
df['Salary']=pd.to_numeric(df['Salary'],errors='coerce')
df[['Age','Salary']]=knn_imputer.fit_transform(df[['Age','Salary']])
df

#
```

```python
df[['ageMissing','salarymissing','citymissing']]=df[['Age','Salary','City']].isnull().astype(int)
df
```

```python
df['missing']=df.isnull().any(axis=1).astype(int)
df
```

```python
#Label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Performance']=le.fit_transform(df['Performance'])
df
#
```

```python
# onehot encoding
df=pd.get_dummies(df,columns=['City'],drop_first=True)
df
```

```python
#ordinal encoding
df['City']=df['City'].map({'Karachi':1,'Islamabad':2,'Peshawar':3,'Lahore':4})
df
```

```python
#targetencoding
target=df.groupby('Age')['Salary'].mean()
df['Age_encoding']=df['Age'].map(target)
df
```

```python
# for scaling features
from sklearn.preprocessing import StandardScaler, RobustScaler
scaler=StandardScaler()
df['Age']=pd.to_numeric(df['Age'],errors='coerce')
df['Salary']=pd.to_numeric(df['Salary'],errors='coerce')
df[['Age','Salary']]=scaler.fit_transform(df[['Age','Salary']])
df
```
